### График разброса значений возраста и потенциала игроков

```python
# Copyright 2020 aaaaaaaalesha

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("CompleteDataset.csv", low_memory=False)

# Распределение по двум значениям.
df.plot.scatter(x="Age", y="Potential", color="purple", marker = 'o', s=40, grid=True, figsize=(16,9))

plt.title("Распределение по значениям возраста и потенциала игроков", fontdict={'fontsize': 16})
plt.xlabel("Возраст", fontdict={'fontsize': 14})
plt.ylabel("Потенциал", fontdict={'fontsize': 14})

plt.xticks(range(15, 49))
plt.yticks(range(45, 100, 5))

plt.savefig('results_viz2.png')
```

### Текстовое описание визуализации

На данном графике можно увидеть распределение(разброс) всех игроков по их возрасту(Age) и потенциалу(Potential) соответственно.

Можно узнать(ответить на вопрос), что:

   - возраст в период с 17 до 34 лет не является существенным показателем показателем потенциала игрока;
   - существуют фтболисты, потенциал которых в 44 года превышает потенциал некоторых футболистов в 21;
