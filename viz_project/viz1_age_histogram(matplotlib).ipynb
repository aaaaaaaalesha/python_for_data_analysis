### Гистограмма распределения возраста игроков в видеоигре FIFA2018

```python
# Copyright 2020 aaaaaaaalesha

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("CompleteDataset.csv", low_memory=False)

age_histogram = df.Age.plot.hist(bins=20, figsize=(16,9), color="yellow", edgecolor="black")

plt.title("Гистограмма распределения возраста игроков", fontdict={'fontsize': 16})
plt.xlabel("Возраст игроков", fontdict={'fontsize': 14})
plt.ylabel("Количество игроков", fontdict={'fontsize': 14})

plt.xticks(range(15, 46))

plt.savefig('results_viz1.png')
```

Данная гистограмма показывает распределение всех футболистов в представденном датафрейме по возрасту.
